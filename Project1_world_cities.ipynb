### Dataset: World Cities Population (5000 rows)

This notebook demonstrates:
1. **pandas-based statistics** (mean, median, mode)  
2. **Standard library implementation** (no pandas/statistics)  
3. **Vertical emoji histogram** of population distribution

Dataset: `world_cities_5000.csv`  
Numeric column analyzed: `population`


## 0) Setup

```

FILE_PATH = "world_cities_5000.csv"
NUMERIC_COLUMN = "population"
CSV_DELIMITER = ","
```

## 1) pandas version â€” compute mean, median, and mode

```

import pandas as pd

df = pd.read_csv(FILE_PATH, delimiter=CSV_DELIMITER)
s = pd.to_numeric(df[NUMERIC_COLUMN], errors="coerce").dropna()

mean_pandas = s.mean()
median_pandas = s.median()
mode_pandas = s.mode().iloc[0] if not s.mode().empty else None

print("pandas results:")
print("  mean  =", mean_pandas)
print("  median=", median_pandas)
print("  mode  =", mode_pandas)
```

## 2) Standard library version â€” compute mean, median, and mode

```

import csv

values = []
with open(FILE_PATH, newline="", encoding="utf-8") as f:
    reader = csv.DictReader(f, delimiter=CSV_DELIMITER)
    for row in reader:
        raw = (row.get(NUMERIC_COLUMN) or "").strip()
        try:
            x = float(raw)
            values.append(x)
        except ValueError:
            pass

n = len(values)
print(f"Parsed {n} numeric values")

mean_stdlib = sum(values) / n if n > 0 else None

values_sorted = sorted(values)
if n == 0:
    median_stdlib = None
elif n % 2 == 1:
    median_stdlib = values_sorted[n // 2]
else:
    median_stdlib = (values_sorted[n // 2 - 1] + values_sorted[n // 2]) / 2

counts = {}
for v in values:
    counts[v] = counts.get(v, 0) + 1
mode_stdlib = max(counts, key=counts.get) if counts else None

print("standard library results:")
print("  mean  =", mean_stdlib)
print("  median=", median_stdlib)
print("  mode  =", mode_stdlib)
```

## 3) Vertical Emoji Histogram (standard library only)

```

# ğŸ“Š Each column = one bin, each ğŸ“ˆ = frequency step

if len(values) == 0:
    print("No values to visualize.")
else:
    num_bins = 10
    vmin, vmax = min(values), max(values)
    span = vmax - vmin
    bins = [0] * num_bins

    for v in values:
        idx = int((v - vmin) / span * (num_bins - 1))
        bins[idx] += 1

    max_count = max(bins)
    height = max_count

    print("ğŸ“Š Vertical Emoji Histogram (Population Distribution) ğŸ“ˆ")
    print()

    for level in range(height, 0, -1):
        row = ""
        for count in bins:
            row += " ğŸ“ˆ " if count >= level else "    "
        print(row)

    print("â”â”" * num_bins)
    for i in range(num_bins):
        print(f"{i+1:02d}", end="  ")
    print()
    print("âœ¨ Each column represents one population bin (lowest â†’ highest) âœ¨")
```
